```python
# 3: Listas - Cadastrar nomes de alunos em uma lista dinamicamente e exibir todos.

def cadastrar_inscritos_colab():
    lista_inscritos = []
    print("Cadastro de Inscritos (Lista).")
    print("Digite 'sair' ou pressione Enter em um campo vazio para finalizar.")

    while True:
        nome = input("Nome do participante: ")

        if nome.lower() == 'sair' or nome == "":
            break

        lista_inscritos.append(nome)
        print(f"'{nome}' cadastrado. Total: {len(lista_inscritos)} inscritos.")

    print("\n---  Lista Final de Inscritos  ---")

    if lista_inscritos:
        for i, inscrito in enumerate(lista_inscritos):
            print(f"{i+1}. {inscrito}")
    else:
        print("Nenhum Participante foi cadastrado.")

cadastrar_inscritos_colab()
```
