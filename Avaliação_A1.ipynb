```python
# 1: Estruturas Condicionais - Simulação de verificação de idade para entrada em eventos.
try:
    idade = int(input("Digite a sua idade: "))

    if idade < 0:
        print("Idade inválida.")
    elif idade < 16:
        print("Entrada Proibida. Evento apenas para maiores de 16 anos.")
    elif idade >= 16 and idade < 18:
        print("Entrada Permitida (16+). Você pode entrar, mas não tem acesso a áreas 18+.")
    else:
        print("Entrada Permitida (18+). Seja bem-vindo(a)!")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```

```python
# 2: Estruturas de Repetição (contadores_pares.py) - Contadores de 1 a 100 com for e while, exibindo apenas números pares.

# ---------------------------------------------
# PARTE 1: Contagem usando FOR
# ---------------------------------------------
print("Contagem de Pares (1 a 100) usando FOR.")
for numero in range(1, 101):
    if numero % 2 == 0:
        print(numero, end=' ')
print("\n--- Fim da contagem FOR ---\n")

# ---------------------------------------------
# PARTE B: Contagem usando WHILE
# ---------------------------------------------
print("Contagem de Pares (1 a 100) usando WHILE.")
numero = 1
while numero <= 100:
    if numero % 2 == 0:
        print(numero, end=' ')

    numero += 1
print("\n--- Fim da contagem WHILE ---")
```

```python
# 3: Listas (cadastro_alunos.py) - Cadastrar nomes de alunos em uma lista dinamicamente e exibir todos.

def cadastrar_alunos_colab():
    lista_alunos = []
    print("Cadastro de Alunos (Lista).")
    print("Digite 'sair' ou pressione Enter em um campo vazio para finalizar.")

    while True:
        nome = input("Nome do aluno: ")

        if nome.lower() == 'sair' or nome == "":
            break

        lista_alunos.append(nome)
        print(f"'{nome}' cadastrado. Total: {len(lista_alunos)} alunos.")

    print("\n---  Lista Final de Alunos (Impressão Correta) ---")

    if lista_alunos:
        for i, aluno in enumerate(lista_alunos):
            print(f"{i+1}. {aluno}")
    else:
        print("Nenhum aluno foi cadastrado.")

cadastrar_alunos_colab()
```

```python
# 4: Dicionários (cadastro_produtos.py) - Sistema simples de cadastro de produtos (nome e preço) em um dicionário.

def cadastrar_produtos_colab():
    cadastro_produtos = {}
    print("Cadastro Simples de Produtos (Dicionário).")
    print("Digite 'sair' no nome para finalizar o cadastro.")

    while True:
        nome_produto = input("\nNome do Produto: ").strip()

        if nome_produto.lower() == 'sair' or not nome_produto:
            break

        try:
            preco = float(input(f"Preço de {nome_produto}: R$ "))

            cadastro_produtos[nome_produto] = preco
            print(f"Produto '{nome_produto}' (R$ {preco:.2f}) cadastrado com sucesso.")

        except ValueError:
            print(" Preço inválido. Digite um número.")
            continue

    print("\n---  Produtos Cadastrados (Recuperação e Exibição) ---")

    if cadastro_produtos:
        for nome, preco in cadastro_produtos.items():
            print(f"Produto: {nome:<20} | Preço: R$ {preco:6.2f}")
    else:
        print("Nenhum produto foi cadastrado.")

cadastrar_produtos_colab()
```
