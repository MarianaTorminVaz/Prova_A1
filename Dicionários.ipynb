```python
# 4: Dicionários - Sistema simples de cadastro de produtos (nome e preço) em um dicionário.

def cadastrar_produtos_colab():
    cadastro_produtos = {}
    print("Cadastro Simples de Produtos (Dicionário).")
    print("Digite 'sair' no nome para finalizar o cadastro.")

    while True:
        nome_produto = input("\nNome do Produto: ").strip()

        if nome_produto.lower() == 'sair' or not nome_produto:
            break

        try:
            preco = float(input(f"Preço de {nome_produto}: R$ "))

            cadastro_produtos[nome_produto] = preco
            print(f"Produto '{nome_produto}' (R$ {preco:.2f}) cadastrado com sucesso.")

        except ValueError:
            print(" Preço inválido. Digite um número.")
            continue

    print("\n---  Produtos Cadastrados ---")

    if cadastro_produtos:
        for nome, preco in cadastro_produtos.items():
            print(f"Produto: {nome:<20} | Preço: R$ {preco:6.2f}")
    else:
        print("Nenhum produto foi cadastrado.")

cadastrar_produtos_colab()
```
