```python
# 1: Estruturas Condicionais - Simulação de verificação de idade para entrada em eventos.
try:
    idade = int(input("Digite a sua idade: "))

    if idade < 0:
        print("Idade inválida.")
    elif idade < 16:
        print("Entrada Proibida. Evento apenas para maiores de 16 anos.")
    elif idade >= 16 and idade < 18:
        print("Entrada Permitida (16+). Você pode entrar, mas não tem acesso a áreas 18+.")
    else:
        print("Entrada Permitida (18+). Seja bem-vindo(a)!")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```
