```python
# 1: Estruturas Condicionais - Simulação de verificação de idade para entrada em eventos.
try:
    idade = int(input("Digite a sua idade: "))

    if idade < 0:
        print("Idade inválida.")
    elif idade < 18:
        print("Entrada Proibida. Evento apenas para maiores de 18 anos.")
    elif idade >= 65:
        print("Entrada Permitida. Acesso prioritário para idosos.")
    else:
        print("Entrada Permitida. Bem-vindo(a) ao evento!")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```
